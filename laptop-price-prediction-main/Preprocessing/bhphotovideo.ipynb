```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns", 100)
pd.set_option("display.max_rows", 100)
```

```python
df = pd.read_csv("../data/raw/raw_bhphotovideo.csv")
df.head(20)
```

```python
df['Name'].nunique()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(df['Name'], bins=20, kde=False)

plt.ylabel('Number')
plt.show()
```

```python
missing_values = df.isnull().sum()

highlighted_columns = ['Name', 'Price', 'CPU', 'GPU', 'Total-Installed-Memory', 'Resolution', 'Total-Install-Capacity', 'Operating-System', 'Capacity', 'Processor',]

plt.figure(figsize=(20, 10))
missing_values.plot(kind='bar', color='lightgray')

for col in highlighted_columns:
    if col in missing_values.index:
        plt.bar(col, missing_values[col], color='green')

plt.show()
```
