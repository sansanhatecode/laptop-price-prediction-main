```python
url_list = defaultdict(list)
prices = []
data_with_price = defaultdict(list)
for i in range(1,401):
    # Get page html sources
    page_url = "https://www.amazon.com/s?k=Laptop+Computer&i=computers&rh=n%3A565108%2Cp_n_graphics_type_browse-bin%3A14292274011&dc&page={}&crid=3P0IV4JZ1OSKW&qid=1702024162&rnid=14292272011&sprefix=laptop%2Caps%2C825&ref=sr_pg_{}".format(i,i)
    browser.get(page_url)
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    
    # Showing progress
    print("Scraping page", str(i))
    
    # Find all product block
    results = soup.find_all("div", {"class": product_div_class}) 
    for item in results:
        price_check = item.find('span', {"class": 'a-price'})
        if price_check:
            url_list['with_price'].append(parent_url + item.find('a', {'class': url_a_class}).get('href'))
            price = price_check.find('span',{"class": price_span}).text.replace('$','').strip()
            if ',' in price:
                prices.append(float(price.replace(',','')))
            else:
                prices.append(float(price))
        else:
            url_list['no_price'].append(parent_url + item.find('a', {'class': url_a_class}).get('href'))
            
# print(len(url_list))   
```

# Get details

```python
def get_spec(details):
#     table = soup2.find('div',{'id': 'productDetails_expanderSectionTables'})

#     lines = table.find_all('tr')
#     categories = []
#     specs = []
#     for line in lines:
#         categories.append(line.find('th').text.strip())
#         specs.append(line.find('td').text.strip())
#     details = {k:v for (k,v) in zip(categories, specs)}

    columns_new = ['Brand','Series','CPU', 'GPU','RAM', 'Monitor','Resolution','Storage','OS','Weight']

    config = {k:None for k in columns_new}

    # Brand
    if 'Brand' in details:
        config['Brand'] = details['Brand']
    # Series 
    if 'Series' in details:
        if 'Model Name' in details:
            config['Series'] = details['Model Name']
        else:
            config['Series'] = details['Series']
    # CPU
    if 'CPU Model Number' in details:
        if ('CPU Manufaturer' in details) and (details['CPU Manufacturer'] in details['CPU Model Number']):
            if 'CPU Speed' in details:
                config['CPU'] = details['CPU Model Number'] + " " + details['CPU Speed']
            else:
                config['CPU'] = details['CPU Model Number']
        else:
            if 'CPU Model' in details:
                if 'CPU Speed' in details:
                    config['CPU'] = details['CPU Manufacturer'] + " " + details['CPU Model'] + " " + details["CPU Model Number"] + " " + details["CPU Speed"]
                else:
                    config['CPU'] = details['CPU Manufacturer'] + " " + details['CPU Model'] + " " + details["CPU Model Number"]
    else:
        if 'CPU Manufacturer' in details:
            cpu_name = ""
            if 'CPU Model' in details:
                if details['CPU Manufacturer'] in details["CPU Model"]:
                    cpu_name = details["CPU Model"]
                else:
                    cpu_name = details['CPU Manufacturer'] + " " + details["CPU Model"]

                if 'CPU Speed' in details:
                    config['CPU'] = cpu_name + " " + details['CPU Speed']
                else:
                    config['CPU'] = cpu_name
            else:
                config['CPU'] = details['CPU Manufacturer']
    # GPU        
    if 'Graphics Coprocessor' in details:
        config['GPU'] = details['Graphics Coprocessor']
    else:
        if 'Chipset Brand' in details:
            config['GPU'] = details['Chipset Brand']
        else:
            config['GPU'] = 'Integrated'

    # RAM
    if 'Ram Memory Installed Size' in details:
        config["RAM"] = details['Ram Memory Installed Size']

    # Monitor
    if "Screen Size" in details:
        config["Monitor"] = details["Screen Size"]
    # Resolution
    if "Screen Resolution" in details:
        config["Resolution"] = details["Screen Resolution"]

    # Storage
    if "Hard Disk Size" in details:
        if "Hard Disk Description" in details:
            config["Storage"] = details["Hard Disk Size"] + " " + details["Hard Disk Description"]
        else:
            config["Storage"] = details["Hard Disk Size"]

    # OS
    if "Operating System" in details:
        config["OS"] = details["Operating System"]
    else:
        config["OS"] = "No OS"

    # Weight
    if "Item Weight" in details:
        config["Weight"] = details["Item Weight"]
    
    return config
```

# Scrape all products that have prices in the url page

```python
def get_spec_with_price(url):
    browser.get(url)
    
    soup2 = BeautifulSoup(browser.page_source, 'html.parser')
    
    try:
        table = soup2.find('div',{'id': 'productDetails_expanderSectionTables'})
        lines = table.find_all('tr')
        categories = []
        specs = []
        for line in lines:
            categories.append(line.find('th').text.strip())
            specs.append(line.find('td').text.strip())
        details = {k:v for (k,v) in zip(categories, specs)}
        return get_spec(details)

    except:
        table = soup2.find('div',{'id': 'productOverview_feature_div'})
        lines = table.find_all('tr')
        categories = []
        specs = []
        for line in lines:
            categories.append(line.find_all('td')[0].text.strip())
            specs.append(line.find_all('td')[1].text.strip())
        details = {k:v for (k,v) in zip(categories, specs)}
        return get_spec(details)
```
