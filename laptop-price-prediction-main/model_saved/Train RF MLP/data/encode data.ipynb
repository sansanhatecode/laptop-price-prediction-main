```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('merge.csv')
data.drop(columns=['CPU Name','GPU Name'],axis=1,inplace=True)
data['Brand'] = data['Brand'].str.lower()
data
```

```python
import pickle
def encode(brand): 
    LabelEncoder_brands = LabelEncoder()
    LabelEncoder_brands.fit(data['Brand'])
    return LabelEncoder_brands.transform([brand])[0]
data['Encoded_Brand'] = data['Brand'].apply(encode)
price_column = data.pop('Price')
data['Price'] = price_column
encoded_brand_dict = dict(zip(data['Brand'],data['Encoded_Brand']))
# Specify the file path to save the dictionary
file_path = 'encoded_brand_dict.pkl'

# Save the dictionary to the file
with open(file_path, 'wb') as file:
    pickle.dump(encoded_brand_dict, file)
    # Specify the file path to load the dictionary
file_path = 'encoded_brand_dict.pkl'
file_path2 = 'encoded_OS_dict.pkl'
def encode2(OS):
    LabelEncoder_OS = LabelEncoder()
    LabelEncoder_OS.fit(data['OS'])
    return LabelEncoder_OS.transform([OS])[0]
data['Encoded_OS'] = data['OS'].apply(encode2)
OS_column = data.pop('OS')
data['OS'] = OS_column
encoded_OS_dict = dict(zip(data['OS'],data['Encoded_OS']))
# Save the dictionary to the file
with open(file_path2, 'wb') as file:
    pickle.dump(encoded_OS_dict, file)
data['Resolution'] = data['Width']*data['Height']
data.drop(columns=['Width','Height'],axis=1,inplace=True)
data
```

```python
data.drop(columns=['Brand','OS'],axis=1,inplace=True)
data.to_csv('preprocess.csv',index=False)
```
