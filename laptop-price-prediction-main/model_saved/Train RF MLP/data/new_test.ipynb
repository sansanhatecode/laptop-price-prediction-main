```python
import pandas as pd
data = pd.read_csv('1.csv')
# Define the condition for removing rows based on the 'Valid' column
condition = data['Valid'] > 0.01

# Apply the condition to filter the DataFrame
data = data[condition]
data.drop(columns=['Unnamed: 0','Unnamed: 14','Valid'], inplace=True)
data.to_csv('merge.csv', index=False)
data
```
